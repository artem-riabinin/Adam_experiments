```python
import matplotlib.pyplot as plt
import re

# Function to parse the log file and extract epoch and train loss values
def parse_log_file(log_file_path):
    epochs = []
    train_losses = []
    
    # Open the log file and read the lines
    with open(log_file_path, 'r') as file:
        lines = file.readlines()
        
        # Regular expression to match the epoch and train loss in each line
        pattern = r"Epoch \[(\d+)/\d+\], Train Loss: ([\d.]+)"
        
        for line in lines:
            match = re.search(pattern, line)
            if match:
                epoch = int(match.group(1))
                train_loss = float(match.group(2))
                epochs.append(epoch)
                train_losses.append(train_loss)
    
    return epochs, train_losses

# Function to plot the train loss vs. epochs for multiple log files
def plot_multiple_train_losses(log_file_paths):
    plt.figure(figsize=(10, 6))
    
    # Iterate over each log file and plot its data
    k = 0
    for log_file_path in log_file_paths:
        epochs, train_losses = parse_log_file(log_file_path)
        if k == 0:
            plt.plot(epochs, train_losses, label=r'Adam: $\beta_1 = 0.1$, $\beta_2 = 0.999$, $\alpha = 0.002$', marker='o', markersize=3, linewidth=1)
        if k == 1:
            plt.plot(epochs, train_losses, label=r'Adam: $\beta_1 = 0.1$, $\beta_2 = 0.5$, $\alpha = 0.002$', marker='o', markersize=3, linewidth=1)
        if k == 2:
            plt.plot(epochs, train_losses, label=r'Adam: $\beta_1 = 0.1$, $\beta_2 = 0.1$, $\alpha = 0.001$', marker='o', markersize=3, linewidth=1)
        k+=1
    
    # Customize the plot
    plt.xlabel('Epoch', fontsize=16)
    plt.ylabel('Train Loss', fontsize=16)
    plt.grid(True)
    plt.legend(fontsize=14)
    plt.savefig('resnet_exp_1.pdf', format='pdf')
    plt.show()

# Main function
if __name__ == "__main__":
    log_file_paths = [
        'output1.txt',  
        'output3.txt',
        'output2.txt'
    ]
    
    # Plot train loss vs epochs for all log files
    plot_multiple_train_losses(log_file_paths)
```

```python
import matplotlib.pyplot as plt
import re
import os

# Function to parse the log file and extract epoch and train loss values
def parse_log_file(log_file_path):
    epochs = []
    train_losses = []
    
    # Open the log file and read the lines
    with open(log_file_path, 'r') as file:
        lines = file.readlines()
        
        # Regular expression to match the epoch and train loss in each line
        pattern = r"Epoch \[(\d+)/\d+\], Train Loss: ([\d.]+)"
        
        for line in lines:
            match = re.search(pattern, line)
            if match:
                epoch = int(match.group(1))
                train_loss = float(match.group(2))
                epochs.append(epoch)
                train_losses.append(train_loss)
    
    return epochs, train_losses

# Function to plot the train loss vs. epochs for multiple log files in two groups
def plot_multiple_train_losses(group1_files, group2_files):
    # Create a figure with 2 subplots side by side
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns
    
    # Plot first group of files in the first subplot
    for log_file_path in group1_files:
        epochs, train_losses = parse_log_file(log_file_path)
        
        # Extract the file name without extension for the label
        file_name = os.path.splitext(os.path.basename(log_file_path))[0]
        
        # Plot the data in the first subplot with specific styles
        axes[0].plot(epochs, train_losses, label=f'Group 1: {file_name}', marker='o', markersize=4, linewidth=1)
    
    # Customize the first subplot
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Train Loss')
    axes[0].set_title('Train Loss vs Epoch for Group 1')
    axes[0].grid(True)
    axes[0].legend()
    
    # Plot second group of files in the second subplot
    for log_file_path in group2_files:
        epochs, train_losses = parse_log_file(log_file_path)
        
        # Extract the file name without extension for the label
        file_name = os.path.splitext(os.path.basename(log_file_path))[0]
        
        # Plot the data in the second subplot with different styles
        axes[1].plot(epochs, train_losses, label=f'Group 2: {file_name}', marker='x', markersize=4, linewidth=1)
    
    # Customize the second subplot
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Train Loss')
    axes[1].set_title('Train Loss vs Epoch for Group 2')
    axes[1].grid(True)
    axes[1].legend()
    
    # Show the plot with both subplots
    plt.tight_layout()  # Automatically adjust subplot parameters for better spacing
    plt.show()

# Main function
if __name__ == "__main__":
    # Define the first group of log files
    group1_files = [
        'output1.txt',  # Replace with actual paths
        'output2.txt'
    ]
    
    # Define the second group of log files
    group2_files = [
        'output1.txt',  # Replace with actual paths
        'output2.txt'
    ]
    
    # Plot train loss vs epochs for both groups of files on separate subplots
    plot_multiple_train_losses(group1_files, group2_files)
```
